# Image Classification Performance Analysis: ResNet vs. MobileNetV2

In this notebook, we compare the performance of two popular convolutional neural networks, ResNet and MobileNetV2, on various image classification tasks. Both models were initially pre-trained and subsequently fine-tuned on our specific dataset. The fine-tuning process involved adjusting the models to better recognize and classify images from our domain-specific collection. ResNet achieved 99% accuracy on the test set, significantly outperforming MobileNetV2, which scored 74%.
